```python
vote = df['votes']

min_value = min(vote)
max_value = max(vote)
min_range = -1
max_range = 1

df['votes'] = (df['votes'] - min_value)/(max_value - min_value)*(max_range - min_range) + min_range
```

```python
df
```

```python
rating = df['rating']

min_value = min(rating)
max_value = max(rating)
min_range = -1
max_range = 1

df['rating'] = (df['rating'] - min_value)/(max_value - min_value)*(max_range - min_range) + min_range
```

```python
df
```

```python
cost = df['approx_cost']

min_value = min(cost)
max_value = max(cost)
min_range = -1
max_range = 1

df['approx_cost'] = (df['approx_cost'] - min_value)/(max_value - min_value)*(max_range - min_range) + min_range
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
```

```python
wcss = []

k_values = range(1,11)

for k in k_values:
    kmeans=KMeans(n_clusters = k)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(k_values, wcss, marker = 'o')
plt.xlabel('No. of Clusters (K)')
plt.ylabel('Within Cluster Sum Of Squares (WCSS)')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3)

kmeans.fit(df)

cluster_labels=kmeans.labels_

cluster_centres = scaler.inverse_transform(kmeans.cluster_centers_)

df['cluster'] = cluster_labels

df = df.assign(cluster = cluster_labels)
```

```python
df
```

```python
plt.scatter(df.cluster, df['S/S'])
plt.xlabel('Clusters')
plt.ylabel('Spending Score')
plt.title('K-Means Clustering')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = df.drop('cluster', axis=1)
```

```python
df
```

```python
dbscan = DBSCAN(eps = 0.5, min_samples = 5)
```

```python
clusters = dbscan.fit_predict(df)
```

```python
df['clusters'] = clusters
```

```python
df
```

```python
plt.scatter(df['S/S'], df['A/I'], c = clusters, cmap = 'viridis' )
plt.xlabel('Spending Score(S/S)')
plt.ylabel('Annual Income(A/I)')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
plt.scatter(df['S/S'], df['Age'], c = clusters, cmap = 'viridis' )
plt.xlabel('Spending Score(S/S)')
plt.ylabel('Age')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
plt.scatter(df['S/S'], df['Gender'], c = clusters, cmap = 'viridis' )
plt.xlabel('Spending Score(S/S)')
plt.ylabel('Gender')
plt.title('DBSCAN Clustering')
plt.show()
```
